```python
#possible_rhythms_list = list(possible_rhythms_set)
def rhythm_column_to_vector(column_text, rhythm_list):

    #Begin by creating a list of 0s, with one 0 per item in the second list. ([0, 0, 0, 0])
    #print("len:%d" %len(rhythm_list))
    output_rhythms = [0]*len(rhythm_list)
    #print(output_rhythms)
    #Take the first element of the first list (call this “item”), and replace the 0 with a 1 in the position of “item” in the second list.
    #Then, move on to the second element. 
    #Repeat the same process that was used to find the match previously. 
    for rhythm in column_text:
        #print(rhythm_list)
        #print(rhythm)
        position_of_rhythm = rhythm_list.index(rhythm)
        #print("position_of_rhythm %d" %position_of_rhythm)
        output_rhythms[position_of_rhythm] = 1
        #print("output_rhythms")
        #print(output_rhythms)
    return output_rhythms
    #print("End function rhythm_column_to_vector")
#rhythm_column_to_vector(["half note"], possible_rhythms_list)
```
